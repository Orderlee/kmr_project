**유동인구수**

```python
floating_population = pd.read_csv("seoulsi_floating_population.csv", encoding='cp949')
floating_population
```

```python
floating_population = floating_population.loc[ :, ['기준_년_코드','상권_코드_명','총_유동인구_수','남성_유동인구_수','여성_유동인구_수']]
floating_population.columns = ['기준년','도로명','총유동인구수','남성유동인구수','여성유동인구수']
floating_population
```

```python
floating_population.info()
```

```python
floating_population['도로명'].str.strip()
floating_population
```

```python
floating_pop2 = floating_population.groupby(['도로명'])['총유동인구수','남성유동인구수','여성유동인구수'].mean().astype(int)
floating_pop2
```

```python
value = floating_pop2['총유동인구수'].values
minimum = min(value)
maximum = max(value)
floating_pop2['총유동인구수'] = floating_pop2['총유동인구수'].transform(func = lambda x : 100*(x - minimum)/(maximum-minimum))
floating_pop2['총유동인구수'] =np.ceil(floating_pop2['총유동인구수']).astype(int)
floating_pop2.head()
```

```python
value = floating_pop2['남성유동인구수'].values
minimum = min(value)
maximum = max(value)
floating_pop2['남성유동인구수'] = floating_pop2['남성유동인구수'].transform(func = lambda x : 100*(x - minimum)/(maximum-minimum))
floating_pop2['남성유동인구수'] =np.ceil(floating_pop2['남성유동인구수']).astype(int)
floating_pop2.head()
```

```python
value = floating_pop2['여성유동인구수'].values
minimum = min(value)
maximum = max(value)
floating_pop2['여성유동인구수'] = floating_pop2['여성유동인구수'].transform(func = lambda x : 100*(x - minimum)/(maximum-minimum))
floating_pop2['여성유동인구수'] =np.ceil(floating_pop2['여성유동인구수']).astype(int)
floating_pop2.head()
```

```python
floating_population = floating_pop2.to_csv('floating_pop_f2.csv',encoding='UTF-8')
```

```python
df_total =pd.read_csv('gwanak_f2.csv',encoding='utf-8',index_col=0)
```

```python
x_data = df_total.iloc[ : , [9,12,15]].values
x_data
```

```python
y_data = df_total.iloc[ : , [8]].values
y_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, Y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error

lin_prediction = lin_reg.predict(X_train)
lin_mse = mean_squared_error(Y_train, lin_prediction)
lin_rmse = np.sqrt(lin_mse)
print('RMSE :', lin_rmse)
```

```python
tree_prediction = tree_reg.predict(X_train)
tree_mse = mean_squared_error(Y_train, tree_prediction)
tree_rmse = np.sqrt(tree_mse)
print('RMSE :', tree_rmse)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, X_train, Y_train,
                         scoring='neg_mean_squared_error', cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(X_train, Y_train)
```

```python
def display_scores(scores):
    print("점수:", scores)
    print("평균:", scores.mean())
    print("표준편차:", scores.std())

def show_scores(model):
    scores = cross_val_score(model, X_train, Y_train.ravel(),
                         scoring='neg_mean_squared_error', cv=5)
    rmse_scores = np.sqrt(-scores)
    display_scores(rmse_scores)
```

```python
show_scores(forest_reg)
show_scores(lin_reg)
```
